```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Initialize NLP tools (run once)
nltk.download('stopwords')
nltk.download('wordnet')
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    if not isinstance(text, str):
        return ""

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Remove HTML tags
    text = re.sub(r'<[^>]+>', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', 'emailaddr', text)

    # Handle JIRA macros
    text = re.sub(r'\{.*?\}', ' ', text)

    # Remove special characters
    text = re.sub(r'[^a-z0-9\s]', ' ', text)

    # Remove XML/JSON snippets
    text = re.sub(r'&lt;|&gt;|&amp;|&quot;', ' ', text)

    # Remove version numbers
    text = re.sub(r'\b\d+\.\d+\.\d+\b', '', text)

    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    # Optional: Remove stopwords
    text = ' '.join([word for word in text.split() if word not in stop_words])

    # Optional: Lemmatization
    text = ' '.join([lemmatizer.lemmatize(word) for word in text.split()])

    return text
```

```python
# Apply preprocessing
df['Processed_text_data'] = df['all_text_data'].apply(preprocess_text)
```

```python
# ✅ STEP 3: Load SBERT Embedder
sbert = SentenceTransformer('all-MiniLM-L6-v2')
def get_embeddings(text_list):
    return sbert.encode(text_list, show_progress_bar=True)
```

```python
# ✅ STEP 4: Stage 1 - Predict Project Lead

X_assignee = get_embeddings(new_df['Processed_text_data'].tolist())
y_assignee = new_df['assignee_encoded'].values

X_train_assignee, X_test_assignee, y_train_assignee, y_test_assignee = train_test_split(
    X_assignee, y_assignee, test_size=0.2, random_state=42)

#lead_class_weights = compute_class_weight('balanced', classes=np.unique(y_train_lead), y=y_train_lead)
clf_assignee = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
clf_assignee.fit(X_train_assignee, y_train_assignee)
```

```python
# prompt: Generate me a code for classification report

import numpy as np
# Ensure assignee_encoder is defined and fitted before using it
# In the provided code, assignee_encoder_new is used for fitting, but assignee_encoder
# is used in the classification report. Let's assume assignee_encoder_new is the correct one.
assignee_encoder = assignee_encoder_new # Or make sure assignee_encoder is defined elsewhere

print("\n🔹 Assignee Classification Report:")
# Predict on the test data
y_pred_assignee = clf_assignee.predict(X_test_assignee)

# Get the unique labels present in the test set
unique_test_labels_assignee = np.unique(y_test_assignee)

# Filter the target names to include only those corresponding to the unique test labels
# Use assignee_encoder.classes_ to map the encoded labels back to original assignee names
# Ensure that the labels used for filtering target names exist in assignee_encoder.classes_
filtered_target_names_assignee = [assignee_encoder.classes_[label] for label in unique_test_labels_assignee if label < len(assignee_encoder.classes_)]

# Generate and print the classification report
print(classification_report(y_test_assignee, y_pred_assignee, target_names=filtered_target_names_assignee, labels=unique_test_labels_assignee))

# Optional: Calculate and print Top-k accuracy (e.g., Top-3 accuracy)
# Check if there are enough classes for Top-k accuracy
if len(np.unique(y_test_assignee)) > 1:
    # Get predicted probabilities for Top-k
    y_pred_proba_assignee = clf_assignee.predict_proba(X_test_assignee)

    # Calculate Top-3 accuracy
    top_k = 3
    # Ensure k is less than or equal to the number of classes
    if top_k > len(assignee_encoder.classes_):
        top_k = len(assignee_encoder.classes_)
        print(f"\nWarning: Adjusted Top-k to {top_k} as it's greater than the number of classes.")

    # Use `labels` argument to match the classes used during training/fitting
    # If your unique_test_labels are a subset of the fitted classes, you might need
    # to align y_test_assignee to the full set of classes if using predict_proba directly
    # without the labels argument in top_k_accuracy_score.
    # However, with predict_proba providing probabilities for all classes fitted by the model,
    # and y_test_assignee containing the true labels, we can directly calculate.
    # Need to ensure the shape of y_pred_proba_assignee and y_test_assignee are compatible.
    # y_test_assignee is the encoded labels. y_pred_proba_assignee has columns corresponding to encoder.classes_.
    # top_k_accuracy_score expects true labels (y_test_assignee) and either predicted labels or probabilities.
    # When providing probabilities, the shape should be (n_samples, n_classes).

    # Calculate Top-k accuracy using the encoded labels for y_true
    # The probas are in the order of clf_assignee.classes_ (which should match assignee_encoder.classes_)
    top_k_acc_assignee = top_k_accuracy_score(y_test_assignee, y_pred_proba_assignee, k=top_k, labels=clf_assignee.classes_)
    print(f"\n🔹 Assignee Top-{top_k} Accuracy: {top_k_acc_assignee:.4f}")
else:
    print("\nNot enough unique classes in the test set to calculate Top-k accuracy.")
```

```python
# prompt: Provide me a inference function for example usage

def predict_assignee(text):
    """
    Predicts the assignee for a given input text using the trained model.

    Args:
        text (str): The input text (e.g., summary, description).

    Returns:
        str: The predicted assignee name.
    """
    # Preprocess the input text
    processed_text = preprocess_text(text)

    # Get embedding for the processed text
    text_embedding = get_embeddings([processed_text]) # get_embeddings expects a list

    # Predict the encoded assignee label
    predicted_encoded_label = clf_assignee.predict(text_embedding)

    # Decode the predicted label back to the original assignee name
    predicted_assignee_grouped = assignee_encoder.inverse_transform(predicted_encoded_label)

    # Return the predicted assignee name (assuming inverse_transform returns an array, take the first element)
    return predicted_assignee_grouped[0]

# Example usage:
input_text = "workflow change shown workflow designer diagram button step reproduce create new project note associated default workflow edit default workflow system copy new one make change workflow text view e g remove transition progress status publish workflow check diagram workflow designer view bug change shown note issue progress status edit workflow indeed see cannot transition workflow actually active see change diagram workflow seems work fine"
predicted_assignee = predict_assignee(input_text)
print(f"Predicted Assignee: {predicted_assignee}")
```

```python
# ✅ STEP 4: Stage 1 - Predict Project Lead

X_lead = get_embeddings(new_df['Processed_text_data'].tolist())
y_lead = new_df['lead_encoded'].values

X_train_lead, X_test_lead, y_train_lead, y_test_lead = train_test_split(
    X_lead, y_lead, test_size=0.2, random_state=42)

#lead_class_weights = compute_class_weight('balanced', classes=np.unique(y_train_lead), y=y_train_lead)
clf_lead = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
clf_lead.fit(X_train_lead, y_train_lead)
```

```python
# ✅ STEP 5: Stage 2 - Predict Assignee using [text + lead_encoded]

X_text_lead_assignee = get_embeddings(new_df['Processed_text_data'].tolist())
lead_encoded = new_df['lead_encoded'].values.reshape(-1, 1)
X_lead_assignee = np.concatenate([X_text_lead_assignee, lead_encoded], axis=1)
y_lead_assignee = new_df['assignee_encoded'].values

X_train_lead_assignee, X_test_lead_assignee, y_train_lead_assignee, y_test_lead_assignee = train_test_split(
    X_lead_assignee, y_lead_assignee, test_size=0.2, random_state=42)

#assignee_class_weights = compute_class_weight('balanced', classes=np.unique(y_train_a), y=y_train_a)
clf_lead_assignee = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
clf_lead_assignee.fit(X_train_lead_assignee, y_train_lead_assignee)
```

```python
# prompt: Generate me a code for classification report

import numpy as np
print("\n🔹 Stage 2 (Text + Lead) Assignee Classification Report:")
y_pred_lead_assignee = clf_lead_assignee.predict(X_test_lead_assignee)

# Get the unique labels present in the test set
unique_test_labels_lead_assignee = np.unique(y_test_lead_assignee)

# Filter the target names to include only those corresponding to the unique test labels
# Use the assignee_encoder (or assignee_encoder_new, ensure consistency)
filtered_target_names_lead_assignee = [assignee_encoder.classes_[label] for label in unique_test_labels_lead_assignee if label < len(assignee_encoder.classes_)]

# Generate and print the classification report
print(classification_report(y_test_lead_assignee, y_pred_lead_assignee, target_names=filtered_target_names_lead_assignee, labels=unique_test_labels_lead_assignee))

# Optional: Calculate and print Top-k accuracy (e.g., Top-3 accuracy)
if len(np.unique(y_test_lead_assignee)) > 1:
    y_pred_proba_lead_assignee = clf_lead_assignee.predict_proba(X_test_lead_assignee)

    top_k = 3
    if top_k > len(assignee_encoder.classes_):
        top_k = len(assignee_encoder.classes_)
        print(f"\nWarning: Adjusted Top-k to {top_k} as it's greater than the number of classes.")

    # Use the class labels the model was trained on for top_k_accuracy_score
    top_k_acc_lead_assignee = top_k_accuracy_score(y_test_lead_assignee, y_pred_proba_lead_assignee, k=top_k, labels=clf_lead_assignee.classes_)
    print(f"\n🔹 Stage 2 (Text + Lead) Assignee Top-{top_k} Accuracy: {top_k_acc_lead_assignee:.4f}")
else:
    print("\nNot enough unique classes in the test set to calculate Top-k accuracy for Stage 2 Assignee prediction.")
```

```python
# prompt: Provide me a inference function for example usage

import numpy as np
def full_prediction_pipeline(text):
    """
    Predicts both the Project Lead and the Assignee for a given input text
    using the trained two-stage model.

    Args:
        text (str): The input text (e.g., summary, description).

    Returns:
        tuple: A tuple containing (predicted_lead, predicted_assignee).
    """
    # Preprocess the input text
    processed_text = preprocess_text(text)

    # Get embedding for the processed text
    text_embedding = get_embeddings([processed_text]) # get_embeddings expects a list

    # Stage 1: Predict Project Lead
    predicted_encoded_lead = clf_lead.predict(text_embedding)
    predicted_lead = lead_encoder.inverse_transform(predicted_encoded_lead)[0]

    # Stage 2: Predict Assignee using text embedding and the predicted lead (encoded)
    # Concatenate text embedding with the predicted lead's encoded value
    # The input to clf_lead_assignee is the text embedding AND the encoded lead.
    # The predicted lead is already encoded in predicted_encoded_lead.
    X_input_stage2 = np.concatenate([text_embedding, predicted_encoded_lead.reshape(-1, 1)], axis=1)

    predicted_encoded_assignee = clf_lead_assignee.predict(X_input_stage2)
    predicted_assignee = assignee_encoder.inverse_transform(predicted_encoded_assignee)[0]

    return predicted_lead, predicted_assignee

# Example usage of the full pipeline:
input_text_example = "Investigate performance issues in the new database query for user profiles."
predicted_lead, predicted_assignee = full_prediction_pipeline(input_text_example)
print(f"\nInput Text: '{input_text_example}'")
print(f"Predicted Project Lead: {predicted_lead}")
print(f"Predicted Assignee: {predicted_assignee}")

# Another example:
input_text_example_2 = "Add new feature request for integrating with the legacy reporting system."
predicted_lead_2, predicted_assignee_2 = full_prediction_pipeline(input_text_example_2)
print(f"\nInput Text: '{input_text_example_2}'")
print(f"Predicted Project Lead: {predicted_lead_2}")
print(f"Predicted Assignee: {predicted_assignee_2}")
```
